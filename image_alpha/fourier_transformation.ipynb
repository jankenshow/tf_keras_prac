```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
x = cv2.imread("data/RPN.png")
# x_gray = cv2.cvtColor(x, cv2.COLOR_BGR2GRAY)


X1 = np.fft.fft2(x[:,:,0])
X1 = np.fft.fftshift(X1)

X1_abs = np.absolute(X1)
X1_abs[X1_abs < 1] = 1  # Assign 1 to elements (<1)
P1 = np.log10(X1_abs)  # Power spectrum

# Normalize to [0, 255]
P1_norm = P1/np.amax(P1)
y1 = np.uint8(np.around(P1_norm*255))


X2 = np.fft.fft2(x[:,:,1])
X2 = np.fft.fftshift(X2)

X2_abs = np.absolute(X2)
X2_abs[X2_abs < 1] = 1  # Assign 1 to elements (<1)
P2 = np.log10(X2_abs)  # Power spectrum

# Normalize to [0, 255]
P2_norm = P2/np.amax(P2)
y2 = np.uint8(np.around(P2_norm*255))


X3 = np.fft.fft2(x[:,:,2])
X3 = np.fft.fftshift(X3)

X3_abs = np.absolute(X3)
X3_abs[X3_abs < 1] = 1  # Assign 1 to elements (<1)
P3 = np.log10(X3_abs)  # Power spectrum

# Normalize to [0, 255]
P3_norm = P3/np.amax(P3)
y3 = np.uint8(np.around(P3_norm*255))
```

```python
x.shape, X1.shape, y1.shape
```

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

ax1.imshow(x)
ax2.imshow(y1, cmap = "gray")
ax3.imshow(y2, cmap = "gray")
ax4.imshow(y3, cmap = "gray")

plt.show()
plt.clf()
```
