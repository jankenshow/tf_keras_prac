### Perceptron

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron
```

```python
iris = load_iris()

X = iris.data[:, (2, 3)] # 花弁の長さ、花弁の幅
y = (iris.target == 0).astype(np.int) # 種類をセトナとそれ以外の分類
```

```python
iris.target_names
```

```python
per_clf = Perceptron(random_state=42)
per_clf.fit(X, y)
```

パーセプトロンの学習アルゴリズムはSGDと似ており、  
scikit-learnのPerceptronクラスは、ハイパーパラメータを  
loss="perceptron", learning_rate="constant", eta0=1(学習率), penalty=None  
に設定したSGDClassifierと同じ

```python
y_pred = per_clf.predict([[2, 0.5]])
print(y_pred)
```

### MLP (DNN)

```python
import numpy as np
from sklearn.datasets import fetch_mldata
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import tensorflow as tf
```

```python
# https://github.com/amplab/datascience-sp14/raw/master/lab7/mldata/mnist-original.mat からデータをダウンロード
# from sklearn.datasets.base import get_data_home 
# print (get_data_home()) # これで表示されるパス/mldata/以下に mnist-original.mat を置く
```

```python
mnist = fetch_mldata('MNIST original', data_home='~/scikit_learn_data/')

mnist_data = mnist.data / 255
mnist_label = mnist.target

# print(type(mnist))
# print(mnist.keys())
# # <class 'sklearn.datasets.base.Bunch'>
# # dict_keys(['DESCR', 'COL_NAMES', 'target', 'data'])
```

```python
train_size = 500
test_size = 100
X_train, X_test, y_train, y_test = train_test_split(
    mnist_data, mnist_label, test_size=test_size, train_size=train_size)
X_train = X_train.astype(np.float32)
X_test = X_test.astype(np.float32)
y_train = y_train.astype(np.int32)
y_test = y_test.astype(np.int32)
```

```python
feature_cols = tf.contrib.learn.infer_real_valued_columns_from_input(X_train)
# feature_cols = tf.feature_column.numeric_column(X_train)
```

```python
feature_cols
```

```python
feature_cols
```

```python
dnn_clf = tf.contrib.learn.DNNClassifier(hidden_units=[300,100], n_classes=10, feature_columns=feature_cols)
dnn_clf = tf.contrib.learn.SKCompat(dnn_clf)
dnn_clf.fit(X_train, y_train, batch_size=50, steps=40000)
```

```python
y_pred = dnn_clf.predict(X_test)
accuracy_score(y_test, y_pred['classes'])
```
